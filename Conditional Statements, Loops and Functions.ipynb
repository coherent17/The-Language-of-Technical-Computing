<h2>Loop (迴圈)</h2>

* **while** loop

    execute a set of `statements` as long as a condition is `True`
    
* **for** loop

    execute a set of statements, once for `each item` **in** a `sequence`
    
* Loop control: `break`, `pass`, `continue`, `exit()`

    * break: terminates the current loop and resumes execution at the next statement (跳脫迴圈)
    
    * pass: do nothing (無作用，當程式還不確定要寫什麼，先放著避免產生語法錯誤，常用在function或loop)
    
    * continue: rejects all the remaining statements in the current iteration of the loop and moves the control back to the top of the loop (本次迴圈continue後的部分直接跳過，進入下一個迴圈)

<h2>While loop</h2>

*  常用在需不斷運作的指令，搭配if-else控制條件的觸發

* Can use `break`, `pass`, `continue`, `exit()` to control loop

<h2>Example</h2>

1. Greatest Common Divisor (GCD) 最大公因數

2. Least Common Multiple (LCM) 最小公倍數

Reference: https://en.wikipedia.org/wiki/Euclidean_algorithm

```python
a, b = 27, 45
ab = a* b
while b != 0:
    a, b = b, a % b
    
print(f"The GCD is {a}")
print(f"The LCM is {ab // a}")
```

<h2>For loop</h2>

*  Used for iterating over a sequence, e.g., `range`, `list`, `tuple`, `set`, `dictionary`, or `string`.

* Can use `break`, `pass`, `continue`, `exit()` to control loop

```python
faang = ['FB', 'AAPL', 'AMZN', 'NFLX', 'GOOGL']

for company in faang:
    if company == 'NFLX':
        break
    print(company)
```

```python
for company in faang:
    if company == 'NFLX':
        print('----- NELX would not be printed -----')
        continue
    print(company)
```

Sign function

$$
\begin{equation}
  sgn(x) :=
    \begin{cases}
      -1 & \text{if $x$ < 0}\\

$$
\begin{equation}
    f(x) = 
    \begin{cases}
        cos(\frac{1}{x}) & \text{$x$ < 0} \\
        sin(\pi x) & \text{0 $\leq$ $x$ $\leq$ 1} \\
        2^{-x} & \text{$x$ > 1}
    \end{cases}
\end{equation}
$$

```python
import math

def f(x):
    if x < 0:
        return math.cos(1/x)
    elif 0 <= x <= 1:
        return math.sin(math.pi * x)
    else:
        return 2 ** (-x)
```

```python
x = list(range(-10, 10, 1))

print(list(map(f, x)))
```

<h2>Arbitrary Arguments, *args</h2>

* If you don't know how many arguments that will be passed into your function, add a `*` before the parameter name in the function definition. (類似解壓縮，依照順序讀取傳入的多個參數)

* 盡量清楚function需要用到哪些參數，避免使用這種方式定義function

```python
def name_list(*args):
    print("The first person is ", args[0])

name_list("Nick", "Emily", "Jackson")
```

```python
def name_list(name1, name2, name3):
    print("The first three persons are {}, {}, and {}.".format(name1, name2, name3))

name = ["Nick", "Emily", "Jackson"]

name_list(*name)
```

<h2>Arbitrary Keyword Arguments, **kwargs</h2>

* If you do not know how many keyword arguments that will be passed into your function, add two asterisk: `**` before the parameter name in the function definition.

* The function will receive a dictionary of arguments, and can access the items accordingly. (在dictionary中，可以透過key去存取value)

* 有較多的自由性傳入參數，已知參數定義完後，最後可以加`**kwargs`收集多傳入的參數，避免error產生


```python
def name_list(**kargs):
    print("The first three persons are {}, {}, and {}.".format(kargs['name1'], kargs['name2'], kargs['name3']))
    
name_list(name1='Nick', name2='Emily', name3='Jackson')
```

* function參數最後沒有加 `**kargs`, 如果有超過預設參數的數量輸入function，將會有error產生!

```python
def name_list(name1, name2, name3):
    print("The first three persons are {}, {}, and {}.".format(name1, name2, name3))
    
name_list(name1='Nick', name2='Emily', name3='Jackson', name4="Tommy")
```

* 有加入`**kargs`

```python
def name_list(name1, name2, name3, **kargs):
    print("The first three persons are {}, {}, and {}.".format(name1, name2, name3))
    
name_list(name1='Nick', name2='Emily', name3='Jackson', name4="Tommy")
```

<h2>Gaussian elimination</h2>

* An algorithm for solving systems of linear equations $Ax=b$

$$
A_{n,n} = \left[\begin{array}{cccc|c}
a_{1,1} & a_{1,2} & \cdots & a_{1,n} & b_{1,1}\\
a_{2,1} & a_{2,2} & \cdots & a_{2,n} & b_{2,1}\\
\vdots  & \vdots  & \ddots & \vdots & \vdots\\
a_{n,1} & a_{n,2} & \cdots & a_{n,n} & b_{n,1}\\
\end{array}\right] =
\left[\begin{array}{ccc|c}
1 & 1 & 1 & 2\\
6 & -4 & 5 & 31\\
5 & 2 & 2 & 13\\
\end{array}\right]
$$

1. **Forward elimination**: reduces a given system to row echelon form

2. **Back substitution**: continues to use row operations until the solution is foun
